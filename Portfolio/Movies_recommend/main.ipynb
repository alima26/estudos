```python
df_imdb_titles = pd.read_csv('./imdb_titles.tsv', sep='\t')
df_imdb_titles.head()
```

```python
df = df_imdb_titles.copy()
```

```python
df.dtypes
```

```python
df.shape
```

```python
# We need only the US titles

selecao = df['region'] == 'US'
```

```python
n1 = df[selecao].shape[0]
n1
```

```python
#creating a new dataframe with region = US

df_us = df[selecao]
```

```python
df_us.head(50)
```

```python
# There are duplicate titles, but the 'ordering' is important. Go to ordenate this column and after keep the 
# first selection of the duplicates
df_us = df_us.sort_values(['titleId', 'ordering'], ascending=[True, True])
df_us.head(50)
```

```python
# Keeping the first titles when are duplicates and removing the others.
df_us = df_us.drop_duplicates(subset='titleId', keep='first')
df_us.head(50)
```
